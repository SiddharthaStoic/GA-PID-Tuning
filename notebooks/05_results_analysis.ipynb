# 05 - Historical Results & Statistical Analysis

## Objective

This notebook analyzes the historical track record of all optimization experiments stored in the database.

The goals are:
- Retrieve all previous optimization runs from SQLite
- Perform statistical analysis on the distribution of PID gains
- Visualize the correlation between Kp, Ki, Kd, and Fitness
- Identify the "Global Best" configuration found across all sessions

```python
import sys, os
sys.path.insert(0, os.path.abspath(".."))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from src.database import ExperimentDB

plt.style.use("seaborn-v0_8-darkgrid")
```

## Database Connection

Access the `experiments.db` file to retrieve historical data.

```python
db = ExperimentDB()

df = db.get_history()

print(f"Total experiments found: {len(df)}")
df.head()
```

## Statistical Summary

Overview of the ranges and averages of the optimized coefficients.

```python
df[['kp', 'ki', 'kd', 'fitness']].describe()
```

## Distribution of PID Gains

Visualize the frequency of the optimized parameters to see where the GA most frequently converges.

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

df['kp'].hist(ax=axes[0], color='skyblue', bins=15)
axes[0].set_title('Kp Distribution')

df['ki'].hist(ax=axes[1], color='salmon', bins=15)
axes[1].set_title('Ki Distribution')

df['kd'].hist(ax=axes[2], color='lightgreen', bins=15)
axes[2].set_title('Kd Distribution')

plt.tight_layout()
plt.show()
```

## Parameter Correlation

Study how the interaction between gains affects final fitness.

```python
plt.figure(figsize=(10, 8))
correlation = df[['kp', 'ki', 'kd', 'fitness']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("PID Parameter Correlation Matrix")
plt.show()
```

## Global Best Configuration

Search the entire database for the absolute best controller ever recorded.

```python
best_ever = db.get_best_ever()

print("--- GLOBAL CHAMPION ---")
print(f"ID        : {best_ever['id']}")
print(f"Timestamp : {best_ever['timestamp']}")
print(f"Kp        : {best_ever['kp']:.4f}")
print(f"Ki        : {best_ever['ki']:.4f}")
print(f"Kd        : {best_ever['kd']:.4f}")
print(f"Fitness   : {best_ever['fitness']:.6f}")
```

## Fitness vs Model Type

Compare if the GA performs differently when optimizing against the Linear vs Identified models.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='model_type', y='fitness', data=df)
plt.title("Optimization Performance by Model Type")
plt.show()
```

This notebook provides a "Big Picture" view of the optimization framework's performance. It confirms that:
1. The GA consistently finds coefficients within a specific range.
2. The database effectively tracks improvements over multiple development sessions.
3. We can statistically defend the choice of the final PID gains based on the aggregate data.
